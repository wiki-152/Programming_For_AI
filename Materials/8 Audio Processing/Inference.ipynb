```python
import IPython
import torchaudio
from tortoise.api import TextToSpeech
```

```python
tts = TextToSpeech()
```

```python
# text = "This audio has been generated by Tortoise Text to Speech Model."
text = "چھوٹا بچہ ہنس رہا ہے۔"
```

```python
# Pick a "preset mode" to determine quality. Options: {"ultra_fast", "fast" (default), "standard", "high_quality"}.
preset = "ultra_fast"
```

```python
gen = tts.tts_with_preset(text, voice_samples=None, conditioning_latents=None, preset=preset)
torchaudio.save('generated.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio('generated.wav')
```

```python
#Cloning Voice From Tortoise Given Samples
from tortoise.utils.audio import load_voice

voice_samples, conditioning_latents = load_voice('angie')

gen1 = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents, preset=preset)
torchaudio.save(f'generated-2.wav', gen1.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-2.wav')
```

# ASR

```python
import whisper

# Model Options
# [tiny, base, small,	medium, large, large-v2]

model = whisper.load_model("base")
result = model.transcribe("generated.wav")
```
