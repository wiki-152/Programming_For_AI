```python
import pandas as pd

file_paths = ["HBL.csv", "ABL.csv", "BOP.csv"]

dfs = []

for file_path in file_paths:
    df = pd.read_csv(file_path)
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)


merged_df.to_csv("merged_data.csv", index=False)
```

```python
import pandas as pd

customer_df=pd.read_csv("Customer.csv")
account_df=pd.read_csv("Account.csv")

# Perform VLOOKUP-like operation using merge
merged_df = customer_df.merge(account_df, on='Customer ID', how='right')

merged_df
```

```python
# Merge only specific columns from both DataFrames
merged_df = customer_df.merge(account_df[['Customer ID', 'Balance (USD)']], on='Customer ID', how='left')

merged_df
```
